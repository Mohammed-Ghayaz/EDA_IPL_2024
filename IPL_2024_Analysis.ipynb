```python
batting_stats = []

for batsman in batsmen:
    bat = df[df['striker']==batsman]
    t = bat['batting_team']
    team = t.iloc[0]
    score = bat['runs_of_bat'].sum()    #Total runs by a batsman
    out = len(df[df['player_dismissed']==batsman])
    no4 = len(bat[bat['runs_of_bat']==4])   #No. of 4s
    no6 = len(bat[bat['runs_of_bat']==6])   #No. of 6s
    average = round(score/out, 2)           #Batting Average
    balls_faced = len(bat[bat['wide']==0])
    strike_rate = round((score/balls_faced)*100, 2)

    batting_stats.append([batsman, team, score, balls_faced, average, strike_rate, no4, no6])

batting_stats
```

```python
bowler_stats = []

for bowler in bowlers:
    ball = df[df['bowler']==bowler]
    wickets = len(ball[(ball['player_dismissed']!=0) & (ball['wicket_type']!='runout')])
    team = ball['bowling_team'].iloc[0]
    r = ball[(ball['byes']==0) & (ball['legbyes']==0)]
    runs_conceded = r['runs_of_bat'].sum() + r['extras'].sum()
    d = ball[(ball['wide']==0) & (ball['noballs']==0)]
    dots = len(d[d['runs_of_bat']==0])
    
    balls_bowled = len(ball[(ball['wide']==0) & (ball['noballs']==0)])
    balls_bowled
    if wickets!=0:
        average = round((runs_conceded/wickets), 2)
        strike_rate = round((balls_bowled/wickets), 2)
    else:
        average = float('inf')
        strike_rate = float('inf')
    overs = str(int(balls_bowled/6)) + "." + str(int(balls_bowled%6))
    overs = float(overs)
    economy = round((runs_conceded/balls_bowled)*6, 2)

    bowler_stats.append([bowler, team, wickets, overs, average, dots, economy, strike_rate])
```

```python
def merge_names(row):
    if row['striker'] > row['non_striker']:
        return f"{row['striker']} & {row['non_striker']}"
    else:
        return f"{row['non_striker']} & {row['striker']}"
```

```python
partnership_df['Partnership'] = df.apply(merge_names, axis=1)
```

```python
partners = partnership_df['Partnership'].unique()
```

```python
partners_df = []
for p in partners:
    out = len(partnership_df[(partnership_df['Partnership']==p) & (partnership_df['player_dismissed']!=0)])
    runs_scored_combined = partnership_df[partnership_df['Partnership']==p]['runs_of_bat'].sum() + partnership_df[partnership_df['Partnership']==p]['extras'].sum()
    if out != 0:
        avg = round((runs_scored_combined/out), 2)
    else:
        avg = float('inf')
    balls_faced = len(partnership_df[(partnership_df['Partnership']==p) & ((partnership_df['wide']==0) & (partnership_df['noballs']==0))])
    overs = str(int(balls_faced/6)) + '.' + str(balls_faced%6)
    overs = float(overs)
    if balls_faced != 0:
        run_rate = round((runs_scored_combined/balls_faced)*6, 2)
    else:
        run_rate = 0
    
    team = partnership_df[partnership_df['Partnership']==p]['batting_team'].iloc[0]
    partners_df.append([p, team, runs_scored_combined, overs, avg, run_rate])
```

```python
partnership_df = pd.DataFrame(partners_df, columns = ['Partnership', 'Team', 'Runs', 'Overs', 'Average', 'Run Rate'])
partnership_df = partnership_df.sort_values('Runs', ascending=False, ignore_index=True)
partnership_df
```

```python
def calc_bdry(row):
    runs_of_bdry = (row['4s'])*4 + (row['6s'])*6
    if row['Runs'] != 0:
        bdry_percent = round((runs_of_bdry/(row['Runs']))*100, 2)
    else:
        bdry_percent = 0
    return bdry_percent
```

```python
import panel as pn
import requests
from PIL import Image
from io import BytesIO


# Function to fetch image from URL
def fetch_image(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses
        img = Image.open(BytesIO(response.content))
        return img
    except requests.exceptions.RequestException as e:
        print(f"Error fetching image from {url}: {e}")
        return None
    except Exception as e:
        print(f"Error processing image from {url}: {e}")
        return None

# Function to update image based on selected team
@pn.depends(select_team1.param.value)
def update_team_image(team):
    team_logo_url = team_logos.get(team, None)
    if team_logo_url:
        img = fetch_image(team_logo_url)
        if img:
            return pn.pane.PNG(img, width=200, height=200, styles={'margin-top': '60px'})
    # Return a placeholder or default image if team logo URL is not found or image fetch fails
    return pn.pane.HTML('<p>Team logo not found or failed to load.</p>')



# Create Panel dashboard layout
dashboard = pn.Column(
    select_team1,
    update_team_image
)

# Display the dashboard
#dashboard.show()
```
